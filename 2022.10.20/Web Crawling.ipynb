```python
import bs4 #웹구조 파싱
import requests #통신
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')  #html파싱
```

```python
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'})  #링크의 키값
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```

```python
lists = items[:4]
for l in lists:
  print(l.text)
```
